````python
df = pd.DataFrame({'A': [1, 2, 2, 3, 4, 5, 5, 5, 6, 7, 7]})
counter = df['A'].count()
res_indices_lst = []
for i in range(0, counter - 1): # goes of rows
    if df['A'][i] == df['A'][i + 1]: # check for row values if equals to the one after it
        res_indices_lst.append(i + 1) # returns row index
df.iloc[res_indices_lst]
````
